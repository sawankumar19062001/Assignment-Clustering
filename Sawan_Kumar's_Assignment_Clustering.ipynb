#1. **What is unsupervised learning in the context of machine learning?**  
> Unsupervised learning is a type of machine learning where the model learns patterns or structures from data without any labeled output.

#2. **How does the K-Means clustering algorithm work?**  
> K-Means is a clustering algorithm that groups data into K clusters based on similarity. Here’s how it works:  
- Pick K random points as "centroids" (cluster centers).  
- Assign each data point to the nearest centroid.  
- Move each centroid to the average position of all points assigned to it.  
- Repeat the assignment and centroid-moving steps until the centroids stop moving or change very little.  
The result is K groups of data points that are close to their respective centroids.


#3. **Explain the concept of a dendrogram in hierarchical clustering.**  
> A dendrogram is a tree-like diagram used in hierarchical clustering to show how data points are grouped. It looks like a family tree, where:  
- Each data point starts as its own cluster (leaf) at the bottom.  
- As the algorithm combines similar clusters, branches form, showing which clusters merge at what point.  



#4. **What is the main difference between K-Means and Hierarchical Clustering?**  
>* K-Means requires you to specify the number of clusters (K) upfront and assigns points to fixed clusters based on distance to centroids.
>* Hierarchical clustering doesn’t need a predefined number of clusters.

#5. **What are the advantages of DBSCAN over K-Means?**  
> DBSCAN has these advantages over K-Means:  
- **No need to specify cluster count**: DBSCAN automatically finds clusters based on data density.  
- **Handles irregular shapes**: It can find non-spherical clusters, unlike K-Means, which assumes spherical clusters.  
- **Identifies noise**: DBSCAN marks outliers (noise points) that don’t belong to any cluster.  



#6. **When would you use Silhouette Score in clustering?**  
> The Silhouette Score is used to evaluate how well data points are clustered. We use it when we want to:  
- Check if points are close to their own cluster and far from other clusters.  
- Compare different clustering models or parameters (e.g., number of clusters in K-Means).  
- Decide the optimal number of clusters by picking the setup with the highest score.  


#7. **What are the limitations of Hierarchical Clustering?**  
> Hierarchical clustering has these limitations:  
- **Slow for large datasets**: It’s computationally expensive, making it impractical for huge datasets.  
- **Sensitive to noise**: Outliers can mess up the merging or splitting process.  
- **Fixed decisions**: Once clusters are merged (or split), the algorithm can’t undo those steps.  



#8. **Why is feature scaling important in clustering algorithms like K-Means?**  
> Feature scaling ensures all features (e.g., height, weight) contribute equally to the clustering process. K-Means uses distances to assign points to clusters, so if features have different scales (e.g., height in meters vs. weight in kilograms), larger-scale features dominate the distance calculations. Scaling (e.g., to a 0-1 range) makes distances fair, leading to better clusters.



#9. **How does DBSCAN identify noise points?**  
> DBSCAN identifies noise points based on density:  
- It looks at each point and checks how many other points are within a distance.  


#10. **Define inertia in the context of K-Means.**  
> Inertia in K-Means is a measure of how tightly packed the clusters are.


#11. **What is the elbow method in K-Means clustering?**  
> The elbow method helps choose the best number of clusters (K) for K-Means.

#12. **Describe the concept of "density" in DBSCAN.**  
> In DBSCAN, density refers to how closely packed data points are in a region.


#13. **Can hierarchical clustering be used on categorical data?**  
> Yes, hierarchical clustering can be used on categorical data.


#14. **What does a negative Silhouette Score indicate?**  
> A negative Silhouette Score means a data point is closer to points in another cluster than to points in its own cluster.


#15. **Explain the term "linkage criteria" in hierarchical clustering.**  
> Linkage criteria determine how the distance between clusters is calculated when merging them in hierarchical clustering.


#16. **Why might K-Means clustering perform poorly on data with varying cluster sizes or density?**  
> K-Means assumes clusters are spherical and roughly equal in size and density.


#17. **What are the core parameters in DBSCAN, and how do they influence clustering?**  
> DBSCAN has two main parameters:  
- **Epsilon**: The maximum distance for points to be considered neighbors. A smaller ε creates tighter clusters; a larger ε merges more points into clusters.  
- **MinPts**: The minimum number of points needed to form a dense region (cluster). Higher MinPts requires denser clusters, while lower MinPts allows looser clusters.



#18. **How does K-Means++ improve upon standard K-Means initialization?**  
> K-Means++ improves the initial placement of centroids in K-Means:  
- Instead of picking random centroids, K-Means++ chooses the first centroid randomly,  
- This spreads out initial centroids, reducing the chance of poor starting points.


#19. **What is agglomerative clustering?**  
> Agglomerative clustering is a type of hierarchical clustering that works bottom-up:  
- Start with each data point as its own cluster.  
- Repeatedly merge the two closest clusters (based on a linkage criterion like single or complete linkage).  



#20. **What makes Silhouette Score a better metric than just inertia for model evaluation?**  
> Silhouette Score is often better than inertia because:  
- **Considers cluster separation**: Silhouette Score measures how close points are to their own cluster vs. other clusters, while inertia only looks at distances within clusters.  
- **Detects misassignments**: A low or negative Silhouette Score shows poor clustering (e.g., points closer to other clusters), which inertia can’t detect.

```python
# 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Visualize the results
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', label='Data points')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Centroids')
plt.title('K-Means Clustering on Synthetic Data (4 Centers)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# 29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Compute silhouette scores for k=2 to 5
sil_scores = []
k_values = range(2, 6)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    y_pred = kmeans.fit_predict(X)
    score = silhouette_score(X, y_pred)
    sil_scores.append(score)

# Plot bar chart
plt.bar(k_values, sil_scores, color='skyblue')
plt.title('Silhouette Scores for K-Means (k=2 to 5)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_values)
plt.show()
```

```python
# 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate overlapping clusters
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=1.5, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
y_pred = kmeans.fit_predict(X)

# Create mesh grid for decision boundaries
h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.contourf(xx, yy, Z, cmap='viridis', alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=50, cmap='viridis')
plt.title('K-Means with Decision Boundaries (Overlapping Clusters)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# 34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot

from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# Compute inertia for k=2 to 6
inertias = []
k_values = range(2, 7)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plot line chart
plt.plot(k_values, inertias, marker='o')
plt.title('Inertia for K-Means (k=2 to 6) on Breast Cancer Data')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.show()
```

```python
# 43. Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN side-by-side

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN

# Generate noisy circular data
X, y_true = make_circles(n_samples=300, factor=0.5, noise=0.05, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=2, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Visualize side-by-side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# K-Means plot
ax1.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
ax1.set_title('K-Means Clustering')
ax1.set_xlabel('Feature 1')
ax1.set_ylabel('Feature 2')

# DBSCAN plot
ax2.scatter(X[y_dbscan != -1, 0], X[y_dbscan != -1, 1], c=y_dbscan[y_dbscan != -1], s=50, cmap='viridis', label='Clustered points')
ax2.scatter(X[y_dbscan == -1, 0], X[y_dbscan == -1, 1], c='red', s=50, marker='x', label='Outliers')
ax2.set_title('DBSCAN Clustering')
ax2.set_xlabel('Feature 1')
ax2.set_ylabel('Feature 2')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
# 44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

# Load Iris dataset
iris = load_iris()
X = iris.data

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
y_pred = kmeans.fit_predict(X)

# Compute silhouette scores
silhouette_vals = silhouette_samples(X, y_pred)
avg_silhouette = silhouette_score(X, y_pred)

# Plot silhouette coefficients
plt.figure(figsize=(8, 6))
y_lower = 10
for i in range(3):
    cluster_silhouette_vals = silhouette_vals[y_pred == i]
    cluster_silhouette_vals.sort()
    size_cluster_i = cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster_i
    plt.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_silhouette_vals, alpha=0.7)
    y_lower = y_upper + 10

plt.axvline(x=avg_silhouette, color='red', linestyle='--', label='Average Silhouette')
plt.title('Silhouette Plot for Iris Dataset (K-Means)')
plt.xlabel('Silhouette Coefficient')
plt.ylabel('Cluster Label')
plt.legend()
plt.show()
```

```python
# 46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans

# Load Wine dataset
wine = load_wine()
X = wine.data[:, :4]  # First 4 features
feature_names = wine.feature_names[:4]

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
y_pred = kmeans.fit_predict(X)

# Create DataFrame
df = pd.DataFrame(X, columns=feature_names)
df['Cluster'] = y_pred

# Plot pairplot
sns.pairplot(df, hue='Cluster', vars=feature_names, palette='viridis')
plt.suptitle('K-Means Clustering on Wine Dataset (First 4 Features)', y=1.02)
plt.show()
```

```python
# 47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count

import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Generate noisy blobs
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_pred = dbscan.fit_predict(X)

# Count clusters and noise
n_clusters = len(np.unique(y_pred[y_pred != -1]))
n_noise = np.sum(y_pred == -1)
print("Number of clusters:", n_clusters)
print("Number of noise points:", n_noise)
```
